```python
## Q4:Write a program that takes a number n as input from the user and generates the first n terms of the series formed 
## by squaring the natural numbers. Sample output Enter a number: 6 The first 6 terms of the series are: 1 4 9 16 25 36
```

```python
num = int(input("Enter a Numder: "))

print("The first", num, "terms of the series are:")

for i in range(1,num,+1):
    print(i*i, end=" ")
```

```python
## Q9:Write a program that prompts the user to enter a number and repeats this process 5 times. 
#The program should accumulate the numbers entered and then display the final running total.
# Sample Output:
## Enter a number: 10
## Enter a number: 15
## Enter a number: 35
## Enter a number: 40
## Enter a number: 50
## The final running total is: 150
```

```python
total = 0

for i in range(5):
    num = int(input("Enter a number: "))
    total = total + num

print("The final running total is:", total)
```

```python
## Write a Python program that:
# 1- Creates a list named clean_countries containing the following countries: [Bangladesh India Pakistan China]

# 2- Repeatedly asks the user to enter a country name.

# 3- Checks whether the entered country exists in the clean_countries list.

# 4-If the country is found in the list:

# Display the message:

# <country_name> is in clean countries list

# Stop the program.

# 5-If the country is not found, keep asking the user to enter another country.

```

```python
clean_countries = ["Pakistan", "Denmark", "Switzerland", "Finland",  "Sweden", "Norway", "Iceland", "New Zealand", "Netherlands", "Luxembourg", "Austria"]
while True:
    country = input("Enter a country name: ")

    if country in clean_countries:
        print(country, "is in clean countries list")
        break
```
